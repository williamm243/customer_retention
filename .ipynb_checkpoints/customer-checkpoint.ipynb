# Business Understanding

Churn is the number of people who stop to use or subscribe to a particular service or product. 

Churn is high in industries where competition is high and consumers have multiple suppliers to choose from

Syriatel a leading mobile network provider in Syria is looking to build a model that will predict the possibilities of a customer churning or remaining in the company. And to know which variables mostly affect the possibility of a customer churning.

This information will then help Syriatel to know which Areas need improvement to avoid the possibilty of a customer leaving which leads to financial losses.

 # <u>OBJECTIVE</u>
 The objective of this project is to build and identify machine learning classification model that will correctly classify if a customer will churn or remain in Syriatel

 ## Other Objectives
- Knowing which variables have great impact on customer churn.
- knowing the relationship between customer churn and other variables.
- Using the features identified to predict future probabilities.

# Data pre processing

```python
# Dropping phone number and state because they are personal identifiers
churn_data = churn_data.drop(['state', 'phone number'], axis=1)
```

```python
# Dropping one variable that has multicollinearlity
churn_data = churn_data.drop(["total day minutes", "total eve minutes", "total night minutes", "total intl minutes"], axis=1)
```

```python
shape_of(churn_data)
```

# Data processing for modeling

```python
# Because churn was either true or false we will convert it to  integer 0 will be false and 1 will be true
churn_data['churn'] = churn_data['churn'].astype(int)
churn_data['churn'].head()
```

```python
# for voice mail because there were only 2 values we will use labelencoder

le = LabelEncoder()
churn_data['international plan'] = le.fit_transform(churn_data['international plan'])
churn_data['voice mail plan'] = le.fit_transform(churn_data['voice mail plan'])

#ohe area code
churn_data = pd.get_dummies(churn_data, columns=['area code'], drop_first=True)

```

```python
shape_of(churn_data)
```

#  Modeling

# preparing data for modeling

```python
# the target variables is churn and all the other remaining columns are predictor variables
y = churn_data['churn']
X = churn_data.drop(['churn'], axis=1)

# split our data for modeling  split it  70training - 30testing to get more testing data because it was imbalanced
X_train, X_test, y_train, y_test = train_test_split(X, y, test_size=0.3, random_state=29)
```

After splitting the data into training and test sets, we use the `starndardScaler()` to fit and transform `X_train` and transform `X_test`. 

-  If you first transform the data and then split into training and test sets this leads to **data leakage**. 

```python
# Scaling data using standardScaler
#instantiate starndardScaler and fit the scaler
scaler = StandardScaler()

X_train_scaled = scaler.fit_transform(X_train)
X_test_scaled = scaler.transform(X_test)
```

```python
# we noticed our data was highly imbalanced 
# let's try to balance it using  SMOTE (Synthetic Minority Oversampling Technique)
smote = SMOTE(random_state=87)

X_train_smote, y_train_smote = smote.fit_resample(X_train_scaled, y_train)
```

# Modelling

## models to build 

- logistic regression model(base model)
- k-Nearest-Neighbors
- Decision tree classifier
- random forest


# Classification Metrics

for this project i choose my classification matrix to  be Recall (Sensitivity or True Positive Rate): 

- Recall measures the proportion of actual churned customers that are correctly identified as churned by the model. In this model it is crucial to capture as many churned customers as possible to take appropriate actions. Therefore, maximizing recall is often important to minimize false negatives churned customers classified as non-churned.

- the aim is to maximize recall

```python
# to reduce redundancy we will use a function to do our modeling

def evaluate_model(model):

    #first fit the  model;
    model.fit(X_train_scaled, y_train)
    # Get the predictions on the test set.
    y_pred = model.predict(X_test_scaled)

    # Get the cross-validation score.
    cross_val_scores = cross_val_score(model, X_train_scaled, y_train, cv=5)

    # confusion matrix.
    cm = ConfusionMatrixDisplay.from_estimator(model, X_train_scaled, y_train);

    #classification report.
    report = print(classification_report(y_test, y_pred))

    return cross_val_scores, cm, report
```

- first model that will act as our baseline model is logistic regression 
- hyperoarameters to be used for our model are:
- penalty (l2) to reduce overfitting
- C (regularization strength)
- random state of 29 for repdrocubility

```python

param_grid = {
    'C': [0.1, 1.0, 10.0],  # Varying the inverse of regularization strength
    'penalty': ['l1', 'l2'],  # Trying different penalty terms
    'solver': ['liblinear', 'saga']  # Trying different solvers
}

# Create the Logistic Regression classifier
lr = LogisticRegression(random_state=29)
base_model = GridSearchCV(lr, param_grid)
```

```python

evaluate_model(base_model)
```

- second model  is the knearestneighborsclassifier 
- hyperparameters for this model are
- 'n_neighbors': [3, 5, 7],  # Varying the number of neighbors (k)
-  'weights': ['uniform', 'distance'],  # Trying different weighting schemes
-   'metric': ['euclidean', 'manhattan']  # Trying different distance metrics
-   random state of 29 for repdrocubility

```python
param_grid = {
    'n_neighbors': [3, 5, 7],  
    'weights': ['uniform', 'distance'],  
    'metric': ['euclidean', 'manhattan'] 
}

knn = KNeighborsClassifier()

knn_model = GridSearchCV(knn, param_grid)
```

```python
evaluate_model(knn_model)
```

- third model is decision tree classifier
- hyperparameters to be used for our model are:
- 'criterion': ['gini', 'entropy'],  # Trying different impurity criteria
-   'max_depth': [None, 5, 10, 20],  # Varying the maximum depth of the tree
-   'min_samples_split': [2, 5, 10],  # Trying different minimum samples for split
-    'min_samples_leaf': [1, 2, 4]  # Trying different minimum samples for leaf


```python
param_grid = {
    'criterion': ['gini', 'entropy'],  
    'max_depth': [None, 5, 10, 20],  
    'min_samples_split': [2, 5, 10],  
    'min_samples_leaf': [1, 2, 4]  
}

# Create the Decision Tree classifier
dt = DecisionTreeClassifier(random_state=29)

dt_model = GridSearchCV(dt, param_grid)
```

```python
evaluate_model(dt_model)
```

- final model that will be used is random classifier 
- hyperoarameter to be used for our model is random state of 29

```python
evaluate_model(RandomForestClassifier(random_state=29))
```

## using smote 

```python
def tuned_model(model):

    model.fit(X_train_smote, y_train_smote)
    y_smote_hat = model.predict(X_test_scaled)
    cross_val_scores = cross_val_score(model, X_train_smote, y_train_smote, cv=5)
    cm = ConfusionMatrixDisplay.from_estimator(model, X_train_smote, y_train_smote)
    report = print(classification_report(y_test, y_smote_hat))

    return cross_val_scores, cm, report
```

```python
#logistic model 
tuned_model(base_model)
```

```python
tuned_model(knn_model)
```

```python
tuned_model(dt_model)
```

```python
tuned_model(RandomForestClassifier(random_state=29))
```

```python
def plot_feature_importances(model, feature_names):
    n_features = len(feature_names)
    plt.figure(figsize=(12, 12))
    plt.barh(range(n_features), model.feature_importances_, align='center')
    plt.yticks(np.arange(n_features), feature_names, fontsize=16)
    plt.xlabel('Feature Importance', fontsize=25)
    plt.ylabel('Feature', fontsize=25)
    plt.title('Feature Importances', fontsize=30)
    plt.show()

# Assuming dt_grid_search is the GridSearchCV object obtained from grid search

# Access the best estimator from GridSearchCV
best_dt = dt_model.best_estimator_

# Plot feature importances
plot_feature_importances(best_dt, X_test.columns)
```
